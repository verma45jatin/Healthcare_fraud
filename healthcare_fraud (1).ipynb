```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Load data
data = pd.read_csv('Healthcare Providers.csv')
data = data.dropna(axis = 0, how ='any')
DropCols = ['index', 'National Provider Identifier',
       'Last Name/Organization Name of the Provider',
       'First Name of the Provider', 'Middle Initial of the Provider','Street Address 1 of the Provider',
       'Street Address 2 of the Provider','Zip Code of the Provider',"HCPCS Code","City of the Provider","HCPCS Description","Place of Service","Credentials of the Provider","Country Code of the Provider","Gender of the Provider","State Code of the Provider"	,"Entity Type of the Provider","State Code of the Provider","Provider Type","Medicare Participation Indicator", "HCPCS Drug Indicator"]
data = data.drop(DropCols, axis = 1)

data.info()
#Cleaning
def RemoveComma(x):
    return x.replace(",","")

data["Average Medicare Allowed Amount"] = pd.to_numeric(data["Average Medicare Allowed Amount"].apply(lambda x: RemoveComma(x)),
                                                             errors= "ignore")
data["Average Submitted Charge Amount"] = pd.to_numeric(data["Average Submitted Charge Amount"].apply(lambda x: RemoveComma(x)),
                                                       errors = "ignore")
data["Average Medicare Payment Amount"] = pd.to_numeric(data["Average Medicare Payment Amount"].apply(lambda x: RemoveComma(x)),
                                                       errors = "ignore")
data["Average Medicare Standardized Amount"] = pd.to_numeric(data["Average Medicare Standardized Amount"].apply(lambda x: RemoveComma(x)),
                                                             errors = "ignore")
col_mean = np.mean(data[["Average Submitted Charge Amount"]])
col_std = np.std(data[["Average Submitted Charge Amount"]])
print(col_mean)
print(col_std)
import category_encoders as ce
from sklearn.preprocessing import StandardScaler

def RemoveComma(x):
    return x.replace(",","")

def Preprocessing(data):



   #2.Binary Encoding.


    BEcols = [var for var in data.columns if data[var].dtype == "O"]

    for col in BEcols:
        encoder = ce.BinaryEncoder(cols = [col])
        dfbin = encoder.fit_transform(data[col])
        data = pd.concat([data,dfbin], axis = 1)
        del data[col]

    #3. One-Hot-Encoding

#     data = pd.get_dummies(data,drop_first = True)



    #4. Standardization

    data_columns = data.columns
    std = StandardScaler()
    data = std.fit_transform(data)
    data = pd.DataFrame(data, columns = data_columns)

    return data


data = Preprocessing(data)
data.head()
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt


# Standardize data
scaler = StandardScaler()
data = scaler.fit_transform(data)

# Perform PCA on the data
pca = PCA(3)
pca.fit(data)
scores = pca.transform(data)

# Calculate the reconstruction error for each data point
reconstructions = pca.inverse_transform(scores)
mse = np.mean(np.square(data - reconstructions), axis=1)

# Find the anomalies based on a threshold value for the reconstruction error
mean = np.mean(mse)
std = np.std(mse)
thresholdpca = mean + 2 * std

anomalies = data[mse > thresholdpca]

import matplotlib.pyplot as plt

# Calculate number of fraudulent and normal claims
n_fraudulent = anomalies.shape[0]
n_normal = data.shape[0] - n_fraudulent

# Create pie chart
plt.pie([n_fraudulent, n_normal], labels=['Fraudulent', 'Normal'], autopct='%1.1f%%')
plt.title('Proportion of Fraudulent and Normal Claims')
plt.show()
print("MSE For PCA : ", mean)

# Create a binary label for each data point indicating whether it is an anomaly or not
y_pred = np.zeros(len(data))
y_pred[mse > thresholdpca] = 1

anomaly_charges = anomalies[:, 1]
normal_charges = data[mse <= thresholdpca, 1]
anomaly_avg = np.mean(anomaly_charges)

anomaly_avg = np.mean(anomaly_charges)*244.861906+105.912666
normal_avg = np.mean(normal_charges)*244.861906+105.912666
print(anomaly_avg)
print(normal_avg)

# Create bar graph
plt.bar(['Anomalies', 'Normal'], [anomaly_avg, normal_avg])
plt.title('Average Submitted Charge Amount for Anomalies vs. Normal Cases')
plt.ylabel('Average Submitted Charge Amount')
plt.show()
anomaly_charges = anomalies[:, 0]
normal_charges = data[mse <= thresholdpca, 0]
anomaly_avg = np.mean(anomaly_charges)

anomaly_avg = np.mean(anomaly_charges)*965.75+353.19
normal_avg = np.mean(normal_charges)*965.75+353.19
print(anomaly_avg)
print(normal_avg)

# Create bar graph
plt.bar(['Anomalies', 'Normal'], [anomaly_avg, normal_avg])
plt.title('Average Medicare Allowed Amount for Anomalies vs. Normal Cases')
plt.ylabel('Average Medicare Allowed Amount')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)
labels = kmeans.labels_
centers = kmeans.cluster_centers_

dists = np.linalg.norm(data - centers[labels], axis=1)
mean = np.mean(dists)
std = np.std(dists)
threshold = mean + 2 * std

anomalies = data[dists > threshold]
anomaly_charges = anomalies[:, 0]
normal_charges = data[mse <= thresholdpca, 0]
anomaly_avg = np.mean(anomaly_charges)

anomaly_avg = np.mean(anomaly_charges)*965.75+353.19
normal_avg = np.mean(normal_charges)*965.75+353.19
print(anomaly_avg)
print(normal_avg)
import matplotlib.pyplot as plt

# Calculate number of fraudulent and normal claims
n_fraudulent = anomalies.shape[0]
n_normal = data.shape[0] - n_fraudulent

# Create pie chart
plt.pie([n_fraudulent, n_normal], labels=['Fraudulent', 'Normal'], autopct='%1.1f%%')
plt.title('Proportion of Fraudulent and Normal Claims')
plt.show()

# Create a binary label for each data point indicating whether it is an anomaly or not
y_pred = np.zeros(len(data))
y_pred[mse > threshold] = 1
labels = kmeans.labels_


# Calculate the silhouette score
score = silhouette_score(data, labels)

print("silhouette score; ",score)
from sklearn.metrics import calinski_harabasz_score

# Assume `data` is your dataset and `labels` are the cluster labels for each data point
score = calinski_harabasz_score(data, labels)

print("calinski_harabasz_score",score)
```
